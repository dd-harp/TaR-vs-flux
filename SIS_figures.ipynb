```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

# Fast Travel, High Beta_2

```python
# Reading in some data:

data_path = "Data/Fast_high.csv"
fast_high = np.zeros((301,5))

with open(data_path) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
            fast_high[line_count] = row
            line_count += 1
    print(f'Processed {line_count} lines.')
    

fast_high = fast_high.transpose()
```

```python
plt.figure(figsize =(8,6))

plt.plot(fast_high[0], fast_high[1], linewidth = 8, color = "b")
plt.plot(fast_high[0], fast_high[2], linewidth = 8, color = "r")

plt.ylim(0,0.9)
# axes label size
plt.tick_params(axis = "both", which = "major", labelsize = 24)
plt.xlabel(r"Transmission Intensity  $\beta_1/\gamma$", fontsize = 24)
#plt.ylabel(r"$I_1$", fontsize = 32, rotation = 0)
plt.ylabel(r"Prevalence  $I^*_1/N_1$", fontsize = 24, rotation = 90)
plt.xticks([0,1,2,3], ["0","1","2","3"])
plt.yticks([0,.25,.5,.75], ["0",".25",".5",".75"])

# legend
plt.legend(["Time at Risk Model","Flux Model"], \
           loc = ("upper left"), fontsize = 24)

plt.tight_layout()

plt.savefig("Figures/Fast_high_patch1_both_2.pdf")
```
